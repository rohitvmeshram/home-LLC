```python
df = pd.DataFrame()
df_bymonth = [df_cs, df_pr1, df_pr2, df_unemp, df_tt, df_nas, df_msa,df_hnf,df_nas,df_hou,df_civ,df_emr,df_pio,df_com]
for df1 in df_bymonth:
    df1["DATE"] = pd.to_datetime(df1["DATE"])
    df1 = df1.set_index("DATE")
    df = pd.concat([df,df1], axis = 1)
print(df.shape)
df.tail()







```

```python
df.isna().sum()
```

```python
mean_values = df.mean().iloc[0]  # Calculate the mode values for each column

# Replace null values in the DataFrame with the respective column mode
df_filled = df.fillna(mean_values)
```
